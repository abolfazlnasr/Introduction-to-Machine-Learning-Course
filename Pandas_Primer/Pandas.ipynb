<div dir="rtl" style="font-size:22px;">
Concatenation (الحاق)
</div>

```python

df1 = pd.DataFrame({
    'نام': ['علی', 'رضا'],
    'سن': [25, 30]
})

df2 = pd.DataFrame({
    'نام': ['مینا', 'سارا'],
    'سن': [22, 28]
})

# الحاق دیتافریم ها به صورت عمودی
df_combined = pd.concat([df1, df2], axis=0, ignore_index=True)
df_combined
```

```python
df1 = pd.DataFrame({
    'نام': ['علی', 'رضا'],
    'سن': [25, 30]
})

df2 = pd.DataFrame({
    'شغل': ['مهندس', 'پزشک']
})

# الحاق دیتافریم ها به صورت افقی
df_combined = pd.concat([df1, df2], axis=1)
df_combined
```

<div dir="rtl" style="font-size:22px;">
Merge (ترکیب بر اساس کلید)
</div>

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'نام': ['علی', 'رضا', 'مینا']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'سن': [25, 30, 22]
})

df_merged = pd.merge(df1, df2, on='ID', how='inner')
df_merged
```

<div dir="rtl" style="font-size:22px;">
Join (اتصال)
</div>

```python
df1 = pd.DataFrame({
    'نام': ['علی', 'رضا', 'مینا'],
    'سن': [25, 30, 22]
}, index=['A', 'B', 'C'])

df2 = pd.DataFrame({
    'شغل': ['مهندس', 'پزشک', 'دانشجو']
}, index=['A', 'B', 'D'])

df_joined = df1.join(df2, how='inner')
df_joined
```

<div dir="rtl" style="font-size:20px;">

توضیحات اضافی:

join: برای اتصال DataFrameها بر اساس اندیس استفاده می‌شود.

merge: برای اتصال DataFrameها بر اساس ستون‌های مشخص.

concat: برای الحاق DataFrameها به صورت عمودی یا افقی.

```python
cars_dictionary = {
    "model" :["Peykan", "Peykan", "Peykan", "BMW X6", "BMW X6", "BMW X6", "Nissan Junior", "Nissan Junior", "Nissan Junior", "Nissan Junior"],
    "Color" :["Red", "White", "Olive",  "Black", "Ametrine", "navy", "Blue", "White", "Blue", "Red"],
    "Weight (KG)" :[990, 1000, 990, 2320, 2320, 2220, 1540, 1540, 1600, 1540],
    "year" :[1346, 1357, 1388, 1391, 1393, 1393, 1375, 1383, 1375, 1389],
    "price" :[25, 30, 48, 2200, 4800, 4800, 105, 105, 130, 145]
}

df = pd.DataFrame(cars_dictionary)
df
```

```python
grouped = df.groupby('model')
type(grouped)
```

<div dir="rtl" style="font-size:20px;">

به‌طور کلی توابعی که می‌توان براساس گروه‌بندی اعمال کرد، به سه دسته‌ی زیر تقسیم می‌شوند:

تجمیع (Aggregation): این توابع، برای تمامی اعضای هر گروه، تنها یک مقدار را به‌دست آورده و به کلید آن نسبت می‌دهند. مانند میانگین، مجموع و غیره.

تبدیل (Transform): این توابع، تغییراتی را با توجه به دیگر داده‌های دسته روی هر داده اعمال می‌کنند. مانند نرمال‌سازی درون دسته‌ای.

فیلتر (Filter): حذف برخی داده‌ها با توجه به اطلاعات دیگر اعضای گروه. مثلاً حذف داده‌هایی که مقداری کمتر از میانگین دارند.
</div>

<div dir="rtl" style="font-size:22px;">
تبدیل
</div>

```python
cars_dictionary = {
    "model" :["Peykan", "Peykan", "Peykan", "BMW X6", "BMW X6", "BMW X6", "Nissan Junior", "Nissan Junior", "Nissan Junior", "Nissan Junior"],
    "Color" :["Red", "White", "Olive",  "Black", "Ametrine", "navy", "Blue", "White", "Blue", "Red"],
    "Weight (KG)" :[990, 1000, 990, 2320, 2320, 2220, 1540, 1540, 1600, 1540],
    "year" :[1346, 1357, 1388, 1391, 1393, 1393, 1375, 1383, 1375, 1389],
    "price" :[25, 30, 48, 2200, 4800, 4800, 105, 105, 130, 145]
}

df = pd.DataFrame(cars_dictionary)
df
```

```python
df_test = df.copy()

peykan_price = df_test.loc[df_test.model == 'Peykan', 'price']
df_test.loc[df_test.model == 'Peykan', 'percent'] = (peykan_price - peykan_price.min()) / (peykan_price.max() - peykan_price.min()) * 100

df_test
```

```python

BMW_price = df_test.loc[df_test.model == 'BMW X6', 'price']
df_test.loc[df_test.model == 'BMW X6', 'percent'] = (BMW_price - BMW_price.min())/(BMW_price.max() - BMW_price.min())*100

Nissan_price = df_test.loc[df_test.model == 'Nissan Junior', 'price']
df_test.loc[df_test.model == 'Nissan Junior', 'percent'] = (Nissan_price - Nissan_price.min())/(Nissan_price.max() - Nissan_price.min())*100

df_test
```
