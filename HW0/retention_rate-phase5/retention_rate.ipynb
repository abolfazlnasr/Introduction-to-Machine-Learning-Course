<b>ویژگی‌های نمودار</b>
</font>
</h3>

<center>
<div dir=rtl style="direction: rtl;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>

| <b style="color:#0099cc">ویژگی</b> | <b style="color:#0099cc">مقدار</b> |
| :---: | :---: |
| اندازه‌ی شکل | ۱۵ در ۸ اینچ |
| بازه اعداد برای تابع نگاشت رنگ | ۰ تا ۵۰ |
| کد فرمت مورد استفاده | <code dir=ltr>.0f</code> |
| تابع نگاشت رنگ (<code>cmap</code>) | <code>BuGn</code> |

</font>
</div>
</center>


```python
df_cp = df.copy()

df_cp['InvoiceDate'] = df_cp['InvoiceDate'].astype(str).str[:7] + '-01'

months_labels = df_cp['InvoiceDate'].unique()

customers_of_months = df_cp.groupby('InvoiceDate')['CustomerId'].unique()

first_time_customers_of_month = df_cp.drop_duplicates(subset='CustomerId').groupby('InvoiceDate')['CustomerId'].unique()
```

```python
cohort_matrix = np.full([len(months_labels), len(months_labels)], np.nan)

for i, month_1 in enumerate(months_labels):
    month_i_customers = first_time_customers_of_month[month_1]

    for j in range(len(months_labels) - i):
        month_j_customers = customers_of_months[months_labels[j+i]]
        cohort_matrix[i][j] = (len(np.intersect1d(month_i_customers, month_j_customers)) / len(month_i_customers)) * 100

cohort_matrix = cohort_matrix.round(2)
```

```python
plt.figure(figsize=(15, 8))

heatmap = sns.heatmap(cohort_matrix, annot=True, fmt=".0f", cmap="BuGn", vmin=0, vmax=50,
                      yticklabels=months_labels,
                      xticklabels=range(1, len(months_labels) + 1)
                      )

plt.title("Retention rates")
plt.xlabel('CohortIndex')
plt.grid(visible=False)
plt.yticks(rotation=0)
plt.tick_params(left=False, bottom=False)

plt.show()
```

<h3>این نمودار از لحاظ مفهومی چه چیزی را نشان می‌دهد؟</h3>
<p>فرض کنید که شما و اشخاص دیگری، در فروردین ماه امسال (بدون توجه به شماره‌ی روز آن) برای نخستین‌بار از یک شرکت خرید انجام داده‌اید. اکنون این شرکت قصد دارد بداند که در ماه‌های بعدی پس از اولین خرید، چند درصد از شما و افرادی مثل شما، مجدداً از شرکت خرید کرده‌اند. به‌عنوان مثال در ماه اردیبهشت ۴۰ درصد شما مجدد از شرکت خرید کرده‌اند، در خرداد ۳۰ درصد، در تیر ۵۰ درصد و الی آخر.</p>
<p>مثالی که آورده شد برای مشتریانی بود که اولین خریدشان را در ماه فروردین انجام داده‌اند یعنی تنها یک ردیف از شکل. حال باید به ازای هرکدام از ماه‌ها (محور عمودی) همین کار را تکرار کنیم.</p>

<h3>به نکات زیر دقت کنید:</h3>
<p>نکته‌ی مهم این است که اگر به تصویر داده‌شده توجه کرده باشید مشاهده خواهید کرد که به شکل یک ماتریس بالامثلثی است و زیر قطر اصلی آن خالی است. این خالی بودن به معنی <code>NaN</code> است و نه <code>0</code>. بنابراین در راه‌حل خود مطمئن شوید که برای این قسمت‌ها مقدار <code>0</code> تعریف نکرده باشید،‌ بلکه نباید مقداری داشته باشند (<code>NaN</code> باشند).</p>
<p>علاوه بر این، ترجیحاً به کمک <code>round(2)</code> اعداد ماتریس خود را تا دو رقم اعشار رُند کنید.</p>

<p>توجه داشته باشید که طبق تصویر داده‌شده، نمودار شما نباید در محور عمودی عنوانی داشته باشد.</p>
