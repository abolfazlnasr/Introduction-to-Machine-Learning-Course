<p dir=rtl align=right style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
<code>legend</code> این نمودار کمی با نمونه‌هایی که پیش از این دیدید فرق دارد. در اینجا، <code>legend</code> استفاده‌شده دارای عنوان و قاب می‌باشد. عنوان <code>legend</code> با استفاده از پارامتر  <code>title</code> انتخاب می‌شود. با <code>True</code> کردن پارامتر <code>frameon</code> قاب لازم برای <code>legend</code> نمایش داده می‌شود.
برای تغییر اندازه‌ی فونت عنوان و رنگ زمینه‌ی قاب می‌توانید از پارامترهای <code>title_fontsize</code> و <code>facecolor</code> استفاده کنید.
</font>
</p>

<center>
<div dir=rtl style="direction: rtl;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>

| :---: | :---: |
| رنگ زمینه‌ی قاب | <code>silver</code> |
| اندازه‌ی فونت عنوان | ۱۶ |
| اندازه‌ی فونت متن‌ها | ۱۴ |

</font>
</div>

```python
df = df[df['Country'] != 'United Kingdom']

country_stats = df.groupby('Country').agg(
    UniqueCustomers=('CustomerId', 'nunique'),
    SalesVolume=('UnitPrice', lambda x: (x * df['Quantity']).sum())
)

qSA = country_stats['SalesVolume'].quantile(0.75)
qCC = country_stats['UniqueCustomers'].quantile(0.75)

country_stats['ConditionGroup'] = country_stats.apply(
    lambda row: 'Highest customer & revenue' if row['UniqueCustomers'] > qCC and row['SalesVolume'] > qSA else
                'High customer & low revenue' if row['UniqueCustomers'] > qCC and row['SalesVolume'] < qSA else
                'High revenue & low customer' if row['UniqueCustomers'] < qCC and row['SalesVolume'] > qSA else
                'Low customer & revenue',
    axis=1
)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

colors = {
    'Highest customer & revenue': 'red',
    'High customer & low revenue': 'blue',
    'High revenue & low customer': 'green',
    'Low customer & revenue': 'cyan'
}

scatter_plots = [plt.scatter([], [], color=color, label=group) for group, color in colors.items()]
ax.scatter(country_stats['SalesVolume'], country_stats['UniqueCustomers'], c=country_stats['ConditionGroup'].map(colors), s=20)

ax.set_xlabel('Sales amount', fontsize=16)
ax.set_ylabel('Number of Customers', fontsize=16)
ax.set_title('Market Study', fontsize=18)

ax.grid(True)
ax.legend(title='Group', handles=scatter_plots, frameon=True, title_fontsize=16, fontsize=14, facecolor='silver')

plt.style.use('fivethirtyeight')

plt.text(country_stats.loc['Spain']['SalesVolume'], country_stats.loc['Spain']['UniqueCustomers'] + 0.5, 'Spain')
plt.text(country_stats.loc['France']['SalesVolume'], country_stats.loc['France']['UniqueCustomers'] + 0.5, 'France')
plt.text(country_stats.loc['Netherlands']['SalesVolume'], country_stats.loc['Netherlands']['UniqueCustomers'] + 0.5, 'Netherlands')

plt.show()
```

<h3>لطفا به نکات زیر دقت کنید:</h3>

<p>منظور از داده‌ها همان مقادیری‌ست که برای تعداد مشتریان و میزان فروش محاسبه کرده‌اید. در صورتی‌که نمره‌‎ی این بخش را کسب نمی‌کنید یعنی که محاسبات شما نادرست است.</p>
<p>برای محاسبه‌ی میزان فروش توجه داشته باشید که باید از ضرب دو ستون <code>Quantity</code> و <code>UnitPrice</code> استفاده کنید و مجموع آن را به ازای هر کشور محاسبه کنید.</p>

<p>برای محاسبه‌ی تعداد مشتریان یکتای هر کشور توجه داشته باشید که نباید از <code>drop_duplicates</code> استفاده کنید زیرا که ممکن است یک مشتری در چند کشور خرید کرده باشد. در این قسمت ابتدا داده‌ها را براساس کشور گروه‌بندی کنید سپس به کمک <code>nunique</code> تعداد موارد یکتا در ستون <code>CustomerId</code> را شمارش کنید.</p>

<p>علاوه بر این، فراموش نکنید که باید کشور پادشاهی متحد (United Kingdom) را از داده‌های خود حذف کنید.</p>
