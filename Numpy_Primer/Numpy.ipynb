ترکیب و تقسیم آرایه‌ها

ترکیب

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# ترکیب عمودی
v_combined = np.vstack((a, b))
print("ترکیب عمودی:\n", v_combined)

# ترکیب افقی
h_combined = np.hstack((a, b))
print("ترکیب افقی:\n", h_combined)
```

تقسیم

```python
arr = np.arange(1, 7).reshape(2, 3)
print("آرایه اصلی:\n", arr)

# تقسیم افقی
left, mid, right = np.hsplit(arr, 3)
print("بخش چپ:\n", left)
print("بخش میانی:\n", mid)
print("بخش راست:\n", right)

# تقسیم عمودی
top, bottom = np.vsplit(arr, 2)
print("بخش بالا:\n", top)
print("بخش پایین:\n", bottom)
```

**توابع جبر خطی**

محاسبه ضرب داخلی دو آرایه

```python
a = np.array([1, 2])
b = np.array([3, 4])
dot_product = np.dot(a, b)
print("ضرب داخلی:", dot_product)
```

ضرب ماتریسی دو آرایه

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A)
print('-------------------------')
print(B)
print('-------------------------')
matrix_product = np.matmul(A, B)
print("ضرب ماتریسی:\n", matrix_product)

matrix_product = A @ B
print("ضرب ماتریسی با @:\n", matrix_product)
```

محاسبه معکوس ماتریس

```python
A = np.array([[1, 2],
              [3, 4]])
inverse_A = np.linalg.inv(A)
print("معکوس ماتریس A:\n", inverse_A)
# Idendity Matrix
print(A @ inverse_A)
```

محاسبه دترمینان ماتریس

```python
det_A = np.linalg.det(A)
print("دترمینان ماتریس A:", det_A)
```

یافتن مقادیر و بردارهای ویژه ماتریس

```python
eig_values, eig_vectors = np.linalg.eig(A)
print("مقادیر ویژه:", eig_values)
print("بردارهای ویژه:\n", eig_vectors)
```

**توابع منطقی و مقایسه‌ای**

انتخاب عناصر بر اساس شرط مشخص

```python
arr = np.array([10, 15, 20, 25, 30])
condition = arr > 20
filtered_arr = np.where(condition, arr, -1)
print("آرایه فیلتر شده:\n", filtered_arr)
```

```python
temp = np.array([[10, 12, 11, 17, 15, 11, 14],
                 [13, 15, 18, 21, 19, 18, 17],
                 [18, 20, 21, 19, 23, 24, 22],
                 [20, 23, 19, 21, 25, 22, 21]])
calendar = np.arange(4*7).reshape(4, 7)
holidays = (calendar%7 >= 5) # آرایه بولی متناظر روزهای آخر هفته (پنج‌شنبه و جمعه)
first_mid = (calendar < 2*7) # آرایه بولی متناظر روزهای دو هفته اول اردیبهشت
print('میانه دمای روزهای آخرِ هفته کل ماه:', np.median(temp[holidays]))
print('میانه دمای دو هفته نخست ماه:', np.median(temp[first_mid]))
print('میانگین دمای روزهای آخرِ هفته در دو هفته پایانی ماه:', np.mean(temp[~first_mid & holidays]))
```
