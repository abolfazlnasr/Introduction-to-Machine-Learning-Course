```python
# pip istall matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<div dir="rtl" style="font-size:22px;">
نمودار خطی (Line Plot)
</div>

```python
# داده‌های نمونه
x = np.linspace(0, 10, 100)
y = np.sin(x)

# ایجاد نمودار خطی
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=5)
plt.title('Line Plot of Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# تغییر سبک به 'ggplot'
plt.style.use('ggplot')

# داده‌های نمونه
x = np.linspace(0, 10, 100)
y = np.cos(x)

# ایجاد نمودار خطی
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Cosine Wave', color='green', linestyle='--', linewidth=2)
plt.title('Line Plot of Cosine Function with ggplot Style')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()
plt.show()
```

<div dir="rtl" style="font-size:22px;">
استفاده از Subplots
</div>

```python
# داده‌های نمونه
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# ایجاد Subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

# نمودار اول
axes[0].plot(x, y1, color='blue', label='Sine Wave')
axes[0].set_title('Sine Function')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].legend()
axes[0].grid(True)

# نمودار دوم
axes[1].plot(x, y2, color='green', label='Cosine Wave')
axes[1].set_title('Cosine Function')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# داده‌های نمونه
x = np.linspace(0, 10, 100)
y = np.sin(x)

# ایجاد نمودار خطی
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=2)
plt.title('Line Plot with Annotations')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)

# اضافه کردن متن به نمودار
plt.text(2, 0.5, 'Peak Point', fontsize=12, color='red')

# اضافه کردن پیکان به نقطه خاصی
plt.annotate('First Peak', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 0.8),
             arrowprops=dict(facecolor='black'))

plt.show()
```

```python
# داده‌های نمونه
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 17, 35, 29]

# ایجاد نمودار میله‌ای
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# اضافه کردن مقدار بالای هر میله
for i, v in enumerate(values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.show()
```

<div dir="rtl" style="font-size:22px;">
نمودار هیستوگرام (Histogram)
</div>

```python
# داده‌های نمونه
data = np.random.randn(1000)

# ایجاد نمودار هیستوگرام
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# داده‌های نمونه
data = np.random.normal(loc=0, scale=1, size=1000)

# ایجاد نمودار هیستوگرام با تابع چگالی
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=100, density=True, alpha=0.6, color='g', edgecolor='black')

# اضافه کردن تابع چگالی نرمال
mu, sigma = 0, 1
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')

plt.title('Histogram and Density Function')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

<div dir="rtl" style="font-size:22px;">
نمودار جعبه‌ای (Box Plot)
</div>

```python
# داده‌های نمونه
data = [np.random.normal(0, std, 1000) for std in range(1, 4)]

# ایجاد نمودار جعبه‌ای
plt.figure(figsize=(10, 6))
plt.boxplot(data, vert=True, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            flierprops=dict(color='red', markeredgecolor='red'))
plt.title('Box Plot Example')
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Std=1', 'Std=2', 'Std=3'])
plt.show()
```

<div dir="rtl" style="font-size:22px;">
نمودار دایره ای (Pie Chart)
</div>

```python
# داده‌های نمونه
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # جدا کردن اولین بخش

# ایجاد نمودار دایره ای
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=120)
plt.title('Pie Chart Example')
plt.axis('equal') 
plt.show()
```
